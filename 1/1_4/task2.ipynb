```python
from cmath import exp, log , sqrt, nan
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(-1.2,1.2,0.001)
```

```python
def funcY(x):
    a = 1
    b = -exp((2/3)* log(x).real).real
    c = x**2 - 1

    d = b**2-(4*a*c)

    y1 = nan
    y2 = nan
    if d>0:
        y1 = (-b+(sqrt(d).real))/(2*a)
        y2 = (-b-(sqrt(d).real))/(2*a)
        # print('D>0 x='+str(x)+' y1='+str(y1)+' y2='+str(y2))
    elif d==0:
        y1 = y2 = -(b/(2*a))
        # print('D=0  x='+str(x)+' y1='+str(y1)+' y2='+str(y2))
    # else:
        # print('D<0  x='+str(x)+' y1='+str(y1)+' y2='+str(y2))
    return [y1,y2]
```

```python
plt.grid()
y1 = []
y2 = []
for el in x:
    y = funcY(el)
    y1.append(y[0])
    y2.append(y[1])
plt.plot(x, y1,'b')
plt.plot(x, y2,'r')
plt.show()
```
